```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 读取数据集
data_train = pd.read_csv('../DataSet/titanic/train.csv')
data_test = pd.read_csv('../DataSet/titanic/test.csv')

# 处理数据
data_train = data_train.drop(['PassengerId', 'Name', 'Sex', 'Cabin', 'Embarked', 'Ticket'], axis=1)
data_test = data_test.drop(['PassengerId', 'Name', 'Sex', 'Cabin', 'Embarked', 'Ticket'], axis=1)

# 获取训练集的结果
data_train_result = data_train.loc[:, 'Survived']
data_train = data_train.drop(['Survived'], axis=1)



# 打印数据集的前几行
print(data_train_result.head())
print('------------')
print(data_train.head())
print('------------')
print(data_test.head())
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.fc1 = torch.nn.Linear(5, 4)
        self.fc2 = torch.nn.Linear(4, 1)
        self.sigmoid = torch.nn.Sigmoid()
        self.relu = torch.nn.ReLU()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
```

```python
model = Model()
criterion = torch.nn.BCEWithLogitsLoss(reduction='mean')
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```
